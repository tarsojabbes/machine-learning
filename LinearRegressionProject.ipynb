```python
x = df[["Avg. Session Length", "Time on App", "Time on Website", "Length of Membership", "Yearly Amount Spent"]]
```

```python
y = df["Yearly Amount Spent"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
lm.coef_
```

```python
predictions = lm.predict(x_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
from sklearn import metrics
```

```python
mean_absolute_errors = metrics.mean_absolute_error(y_test, predictions)
mean_squared_errors = metrics.mean_squared_error(y_test, predictions)
squared_mean_squared_errors = np.sqrt(mean_squared_errors)
```
